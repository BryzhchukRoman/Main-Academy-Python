Написати код для виконання наступних дій:

- Згенерувати рядок з усіма малими та великими символами літер та цифрами і зберегти його як список змінних
- Друк першого символу списку
- Друк останнього символу
- Друк 3-го з початку та 3-го кінця символів
- Зробити зріз перших 10 символів
- Друк тільки символів з індексом, який ділиться на 3, без залишку
- Друк тільки цілих значень зі списку
- Друк оберненого списку з символами розташованими у зворотному напрямку за допомогою зрізу
- Перетворити список у рядок з роздільником "-"


```python
import string

print('*'*40)
print('Згенерувати рядок з усіма малими та великими символами літер та цифрами і зберегти його як список змінних: ')
lst = list(string.ascii_letters + string.digits)
print(lst)
print('*'*40)

print("Перший символ: ")
print(lst[0])
print('*'*40)

print("Останній символ: ")
print(lst[-1])
print('*'*40)

print("Третій від початку : ") 
print(lst[2])  
print('*'*40)

print("Третій з кінця: ")
print(lst[-3])    
print('*'*40)

print("Зріз перших 10 символів: ")
print(lst[:10])
print('*'*40)

print("Друк кожного другого символу: ")
print(lst[::2])
print('*'*40)

print("Друк тільки цілих значень зі списку: ")
int_lst = []
for i in lst:
    try:
        int_lst.append(int(i)) # Тут може бути перетворено лише ціле число
    except:
        pass
    
print(int_lst)
print(int_lst[::-1])
print('*'*40)

print("Рядок зі списку з роздільником '-': ")
print("-".join(lst))  
print('*'*40)
```
