```python
"""У школі вирішили набрати три нових математичних класи. 
   Так як заняття з математики у них проходять в один і той же час, 
   було вирішено виділити кабінет для кожного класу і купити до них нові парти. 
   За кожною партою може сидіти не більше двох учнів. 
   Відомо кількість учнів в кожному з трьох класів. 
   Скільки всього потрібно закупити парт щоб їх вистачило на всіх учнів? 
   
   Програма отримує на вхід три натуральних числа: 
   кількість учнів у кожному з трьох класів.
"""
print(40*"*")

a = int(input('Введіть кількість школярів у першому мат. класі: '))
b = int(input('Введіть кількість школярів у другому мат. класі: '))
c = int(input('Введіть кількість школярів у третьому мат. класі: '))

print(40*"*")

print("Потрібно купити парт:")
print((a%2)+(b%2)+(c%2)+(a//2)+(b//2)+(c//2))

print(40*"*")
```
