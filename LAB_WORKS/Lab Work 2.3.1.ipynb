Написати код для виконання наступних дій:

- Створити рядок з усіма малими та великими символами літер та цифрами
- Друк першого символу рядка
- Друк останнього символу
- Надрукуйте 3-й з початку та 3-го з кінця символів
- Зробити зріз перших 8 символів
- Друк тільки символів з індексом, який ділиться на 3, без залишку
- Друк символу середини тексту рядка
- Друк оберненого списку з символами розташованими у зворотному напрямку за допомогою зрізу


```python
import string


print('*'*40)
print("Створити рядок з усіма малими та великими символами літер та цифрами: ")
text = string.ascii_letters + string.digits
print(text)
print('*'*40)

print("Перший символ: ")
print(text[0])
print('*'*40)

print("Останній символ: ")
print(text[-1]) 
print('*'*40)

print("Третій від початку: ")
print(text[2]) 
print('*'*40)

print("Третій з кінця: ")
print(text[-3]) 
print('*'*40)

print("Зріз перших 8 символів: ")
print(text[:8])
print('*'*40)

print("Друк кожного третього символу: ")
print(text[::3])
print('*'*40)

print("Підрахунок першої та останньої межі: ")
print(text[int(len(text)/2):int(len(text)/2)+1])
print('*'*40)

print("Підрахунок індексу символу середини тексту рядка: ")
print(text[int(len(text) / 2 + 0.5)])
print('*'*40)

print("Друк списку оберненого у зворотній бік : ") 
print(text[::-1])
print('*'*40)
```
